```python
import threading
import time

# Define the number of philosophers and forks
num_philosophers = 5
forks = [threading.Semaphore(1) for i in range(num_philosophers)]

# Define the philosopher class
class Philosopher(threading.Thread):
    def __init__(self, index):
        threading.Thread.__init__(self)
        self.index = index

    def run(self):
        while True:
            # Think for a while
            print(f"Philosopher {self.index} is thinking.")
            time.sleep(2)

            # Wait for left fork
            print(f"Philosopher {self.index} is hungry and waiting for left fork.")
            forks[self.index].acquire()

            # Wait for right fork
            print(f"Philosopher {self.index} is waiting for right fork.")
            forks[(self.index + 1) % num_philosophers].acquire()

            # Eat for a while
            print(f"Philosopher {self.index} is eating.")
            time.sleep(2)

            # Release forks
            forks[self.index].release()
            forks[(self.index + 1) % num_philosophers].release()

# Create the philosopher threads
philosophers = [Philosopher(i) for i in range(num_philosophers)]

# Start the philosopher threads
for philosopher in philosophers:
    philosopher.start()

# Wait for the philosopher threads to finish
for philosopher in philosophers:
    philosopher.join()
```
