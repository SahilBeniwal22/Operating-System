```python
import math
import pandas as pd

# Priority Scheduling Pre-emptive
print("Scheduling used: Priority Scheduling Pre-emptive")
n=int(input("Enter number of processes: "))
burst_time=[]
arrival_time=[0]*n
priority=[]
tat=[0]*n
completion_time=[0]*n
time=0
waiting_time=[0]*n


burst_time=[1,7,3,6,5,15,8]
priority=[2,6,3,5,4,10,9]
arrival_time=[0,1,2,3,4,5,15]


print(burst_time)
print(priority)

temp=priority.copy()
remaining_time=burst_time.copy()

time=time + 1
remaining_time[0]=remaining_time[0]-1

iteration=sum(burst_time)

for i in range(0,iteration+n-1):
  m=min(temp[0:time+1])
  index=priority.index(m)
  if(remaining_time[index]==0):
    temp[index]=math.inf
    completion_time[index]=time
    t=completion_time[index]-arrival_time[index]
    tat[index]=t
    w=tat[index]-burst_time[index]
    waiting_time[index]=w
  else:
    time=time+1
    remaining_time[index]=remaining_time[index]-1


df=pd.DataFrame(list(zip(arrival_time, burst_time, priority, completion_time, tat, waiting_time)),
               columns =['Arrival time', 'Burst time','Priority', 'Completion' , 'TAT', 'Waiting time'])

avg_tat=sum(df['TAT'])/n
avg_wt=sum(df['Waiting time'])/n

print(df)
print("Average TAT: ",avg_tat)
print("Average wt:", avg_wt)
```
